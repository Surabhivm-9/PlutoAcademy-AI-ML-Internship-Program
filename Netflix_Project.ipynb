# Netflix Movies and TV Shows Dataset Analysis

## Objective
The objective of this project is to perform Exploratory Data Analysis (EDA) on the Netflix Movies and TV Shows Dataset. The analysis aims to understand content trends, identify patterns, visualize important insights, and provide recommendations based on the findings.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("netflix_titles.csv")

df.head()
```

# Dataset Inspection

In this section, we examine the structure of the dataset including dimensions, columns, data types, missing values and duplicate records.

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

Dataset Summary

1. The dataset contains Netflix Movies and TV Shows.
2. It includes information such as title, director, cast, country, release year, rating and duration.
3. Both movies and television shows are present.
4. Some attributes contain missing values.
5. The dataset is suitable for trend analysis and content exploration.

# Data Cleaning

The dataset contains missing values in several columns. Missing values will be handled appropriately and duplicate records will be removed.

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Fill missing values

df["director"].fillna("Unknown", inplace=True)

df["cast"].fillna("Not Available", inplace=True)

df["country"].fillna("Unknown", inplace=True)

df["rating"].fillna("Unknown", inplace=True)

df["duration"].fillna("Unknown", inplace=True)

print("Remaining Missing Values:")
print(df.isnull().sum())
```

Cleaning Decisions

1. Duplicate rows were removed.
2. Missing director values were replaced with 'Unknown'.
3. Missing cast values were replaced with 'Not Available'.
4. Missing country values were replaced with 'Unknown'.
5. Missing ratings and duration values were filled appropriately.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

Chart 1: Movies vs TV Shows

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.title('Movies vs TV Shows')
plt.show()
```

Chart 2: Top 10 Countries

```python
plt.figure(figsize=(10,5))
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

Chart 3: Top Ratings

```python
plt.figure(figsize=(10,5))
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title('Top Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

Chart 4: Release Year Trend

```python
plt.figure(figsize=(12,5))
df['release_year'].value_counts().sort_index().plot()
plt.title('Release Trend')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

Chart 5: Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(df['release_year'], bins=20)
plt.title('Distribution of Release Years')
plt.show()
```

Chart 6: Pie Chart

```python
content = df['type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(content,
        labels=content.index,
        autopct='%1.1f%%')
plt.title('Movies vs TV Shows Share')
plt.show()
```
